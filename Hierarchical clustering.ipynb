```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
iris.data[:10]
```

```python
iris.target
```

```python
from sklearn.cluster import AgglomerativeClustering

#Hierarchical clustering
#Ward is default linkage algorithm . So we'll start with that
ward=AgglomerativeClustering(n_clusters=3)
ward_pred=ward.fit_predict(iris.data)
```

```python
#Hierarchical clustering using complete linkage
#Create an instance of AgglomerativeClustering with appropriate parameters

complete=AgglomerativeClustering(n_clusters=3,linkage='complete')

#fit and predict
# make AgglomerativeClustering fit and predict the cluster labels

complete_pred=complete.fit_predict(iris.data)

#Hierarchical clustering using average linkage
#Create an instance of AgglomerativeClustering with appropriate parameters

average=AgglomerativeClustering(n_clusters=3,linkage='average')

#fit and predict
# make AgglomerativeClustering fit and predict the cluster labels

average_pred=average.fit_predict(iris.data)
```

```python
from sklearn.metrics import adjusted_rand_score

ward_ar_score=adjusted_rand_score(iris.target,ward_pred)
```

```python
#Calculated adjusted rand score for complete linkage clustering labels

complete_ar_score=adjusted_rand_score(iris.target,complete_pred)

#Calculated adjusted rand score for average linkage clustering labels

average_ar_score=adjusted_rand_score(iris.target,average_pred)
```

```python
print("Scores :\nward-",ward_ar_score,"\nComplete:",complete_ar_score,"\nAverage:",average_ar_score)
```

```python
#The effect of normalization on clustering
```

```python
iris.data[:15]
```

```python
from sklearn import preprocessing
normalized_x = preprocessing.normalize(iris.data)
normalized_x[:15]
```

```python
#Dendrogram visualization with SciFy
```

```python
from scipy.cluster.hierarchy import linkage

#Scify the linkage type, that accepts ward, complete, average and all other values
#Pick the one that resulted in highest adjusted rand score

linkage_type='ward'
linkage_matrix=linkage(normalized_x,linkage_type)
```

```python
from scipy.cluster.hierarchy import dendrogram
import matplotlib.pyplot as plt
plt.figure(figsize=(42,38))


#plot the data using dendrogram()
dendrogram(linkage_matrix)
plt.show()
```

```python
#Visualization with seaborn's clustermap
```

```python
import seaborn as sns

sns.clustermap(normalized_x,figsize=(12,18),method=linkage_type,cmap='viridis')

#Expand figuresize to a value (18,50) if you want too read sample labels
#Draw back is that you'll need to scrolling to observe dendrogram
plt.show()
```
